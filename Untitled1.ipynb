```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# Load dataset from online source (no need to upload)
url = "https://raw.githubusercontent.com/justmarkham/pycon-2016-tutorial/master/data/sms.tsv"
df = pd.read_csv(url, sep='\t', header=None, names=['label', 'message'])
df.head()
```

```python
# Drop duplicates
df = df.drop_duplicates()

# Convert labels: ham -> 1, spam -> 0
df['label'] = df['label'].map({'ham': 1, 'spam': 0})

# Separate features and labels
X = df['message']
y = df['label']
```

```python
sns.countplot(x='label', data=df)
plt.title('Ham vs Spam Count')
plt.xticks([0, 1], ['Spam', 'Ham'])
plt.xlabel('Label')
plt.ylabel('Count')
plt.show()
```

```python
# Split data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)

# TF-IDF Vectorization
vectorizer = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)
X_train_features = vectorizer.fit_transform(X_train)
X_test_features = vectorizer.transform(X_test)
```

```python
model = LogisticRegression()
model.fit(X_train_features, y_train)
```

```python
# Accuracy
train_acc = accuracy_score(y_train, model.predict(X_train_features))
test_acc = accuracy_score(y_test, model.predict(X_test_features))
print(f"Train Accuracy: {train_acc * 100:.2f}%")
print(f"Test Accuracy: {test_acc * 100:.2f}%")

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, model.predict(X_test_features))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Greens',
            xticklabels=['Spam', 'Ham'], yticklabels=['Spam', 'Ham'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# Classification Report
print("Classification Report:")
print(classification_report(y_test, model.predict(X_test_features), target_names=['Spam', 'Ham']))
```

```python
# Prediction function for user input
def predict_message(text):
    input_features = vectorizer.transform([text])
    prediction = model.predict(input_features)
    return "Ham Mail" if prediction[0] == 1 else "Spam Mail"

# Example usage
example_msg = "Congratulations! You've won a luxury vacation package!"
print("Prediction:", predict_message(example_msg))
```
